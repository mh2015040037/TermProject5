```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = "http://sti.kostat.go.kr/site2/pop_learning/learning_003002.html"
req = requests.get(url, verify=False)
res = req.content

soup = BeautifulSoup(res, 'html.parser')

finder_column = soup.find_all('th')
finder_column = [line.get_text().strip() for line in finder_column]

finder_data = soup.find_all('td')
finder_data = [line.get_text().strip() for line in finder_data]

finder_row=[]
for j in range(len(finder_data)):
    if j%4 == 0:
        finder_row.append([finder_data[j],finder_data[j+1],finder_data[j+2],finder_data[j+3]])

df = pd.DataFrame(finder_row, columns=finder_column)


```
